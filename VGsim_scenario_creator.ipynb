**Some basic imports**

```python
import pandas as pd
import numpy as np
import VGsim
```

**VGsim global parameters**

```python
number_of_sites = 1
populations_number = 1
number_of_susceptible_groups = 1
```

**List divide function**

```python
def divide(lst, n):
    p = len(lst) // n
    if len(lst)-p > 0:
        return [lst[:p]] + divide(lst[p:], n-1)
    else:
        return [lst]
```

**Creating Dataframe with segmented statistics of the results of the simulator**

```python
df = pd.DataFrame(columns = ['seg1_count', 'seg1_mean', 'seg1_std', 'seg1_min', 'seg1_25%', 'seg1_50%', 'seg1_75%', 'seg1_max',
                  'seg2_count', 'seg2_mean', 'seg2_std', 'seg2_min', 'seg2_25%', 'seg2_50%', 'seg2_75%', 'seg2_max',
                  'seg3_count', 'seg3_mean', 'seg3_std', 'seg3_min', 'seg3_25%', 'seg3_50%', 'seg3_75%', 'seg3_max',
                  'seg4_count', 'seg4_mean', 'seg4_std', 'seg4_min', 'seg4_25%', 'seg4_50%', 'seg4_75%', 'seg4_max',
                  'transmission_rate', 'recovery_rate'])
```

**Running VGsim with iterable parameters and writing the results into Dataframe**

```python
for i in range(90):
  simulator = VGsim.Simulator(number_of_sites, populations_number, number_of_susceptible_groups, seed=1234)
  simulator.set_transmission_rate(1 + i*0.1)
  for j in range(i+1):
    simulator.set_recovery_rate(j*0.1)
    simulator.set_sampling_rate(1)
    simulator.set_mutation_rate(0, mutation=0)
    rates = pd.Series([(1 + i*0.1), (j*0.1)], index=['transmission_rate', 'recovery_rate'])
    simulator.simulate(1000000)
    Dataset = np.zeros((4, 8))
    time, pop, hap = simulator.sample_data(True)
    time_split = divide(time, 4)
    for k in range(4):
        temp_df = pd.DataFrame({'time': time_split[k]})
        Dataset[k][0] = temp_df.describe(include='all').loc['count']
        Dataset[k][1] = temp_df.describe(include='all').loc['mean']
        Dataset[k][2] = temp_df.describe(include='all').loc['std']
        Dataset[k][3] = temp_df.describe(include='all').loc['min']
        Dataset[k][4] = temp_df.describe(include='all').loc['25%']
        Dataset[k][5] = temp_df.describe(include='all').loc['50%']
        Dataset[k][6] = temp_df.describe(include='all').loc['75%']
        Dataset[k][7] = temp_df.describe(include='all').loc['max']
    temp_series = pd.Series(Dataset.ravel(), index=['seg1_count', 'seg1_mean', 'seg1_std', 'seg1_min', 'seg1_25%', 'seg1_50%', 'seg1_75%', 'seg1_max',
                      'seg2_count', 'seg2_mean', 'seg2_std', 'seg2_min', 'seg2_25%', 'seg2_50%', 'seg2_75%', 'seg2_max',
                      'seg3_count', 'seg3_mean', 'seg3_std', 'seg3_min', 'seg3_25%', 'seg3_50%', 'seg3_75%', 'seg3_max',
                      'seg4_count', 'seg4_mean', 'seg4_std', 'seg4_min', 'seg4_25%', 'seg4_50%', 'seg4_75%', 'seg4_max'])
    temp_series = temp_series.append(rates)
    df = df.append(temp_series, ignore_index=True)
```

**Formating the Dataframe**

```python
df.drop(df.index[:21], inplace=True)
df.reset_index(drop=True, inplace=True)
df.drop(columns=['seg2_count', 'seg3_count', 'seg4_count'], inplace=True)
df['seg1_count'] = df['seg1_count'].apply(lambda x: x*4)
```

**Saving the table as .csv file**

```python
df.to_csv('samples.csv', index=False)
```
